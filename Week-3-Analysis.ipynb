```python
import pandas as pd
```

I am reading in a list of the propositions and committees from the csv file

```python
props = pd.read_csv("http://www.firstpythonnotebook.org/_static/committees.csv")
```

```python
props.head(7)
```

read in the list of contributions to valid measures

```python
contribs = pd.read_csv("http://www.firstpythonnotebook.org/_static/contributions.csv")
```

```python
props.prop_name.value_counts()
```

```python
props.prop_name.value_counts().reset_index()
```

```python
prop67 = props[props.prop_name == "PROPOSITION 067- REFERENDUM TO OVERTURN BAN ON SINGLE-USE PLASTIC BAGS."]
```

```python
prop67.head()
```

```python
prop67.info()
```

```python
contribs.info()
```

```python
merged = pd.merge(prop67, contribs, on="calaccess_committee_id")
```

```python
merged.head()
```

```python
merged.info()
```

```python
merged.amount.sum()
```

```python
merged.committee_position.value_counts()
```

```python
support = merged[merged.committee_position == 'SUPPORT']
```

```python
oppose = merged[merged.committee_position == "OPPOSE"]
```

```python
support.amount.sum()
```

```python
oppose.amount.sum()
```

```python
support.amount.sum() / merged.amount.sum()
```

```python
merged.sort_values("amount", ascending=False)
```

```python
oppose.sort_values("amount", ascending=False).head(10)
```

```python
merged.groupby("committee_name_x").amount.sum().reset_index().sort_values("amount", ascending=False)
```

```python
merged.groupby(["contributor_firstname","contributor_lastname", "committee_position"]).amount.sum().reset_index().sort_values("amount", ascending=False).head(20)
```

```python
merged["contributor_fullname"] = merged["contributor_firstname"].map(str) + " " + merged["contributor_lastname"].map(str)
```

```python
merged["contributor_fullname"] = merged["contributor_fullname"].str.lstrip('nan')
```

```python
merged.groupby(["contributor_fullname", "committee_position"]).amount.sum().reset_index().sort_values("amount", ascending=False).head(20)
```
